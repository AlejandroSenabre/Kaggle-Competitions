```python
#Asignar valor segun nombre
def assign_value(x):
    if "Mr." in x:
        return 1
    elif "Master" in x:
        return 2
    elif "Mrs." in x:
        return 3
    elif "Miss" in x:
        return 4
    elif "Lady" in x or "Sir" in x:
        return 5
    elif "Rev" in x:
        return 6
    elif "Dr" in x:
        return 7
    else:
        return 0
```

```python
#Aplico la función anterior segun apodo
df['Title']=df['Name'].apply(assign_value)

#Aplico a Test
df_test['Title']=df_test['Name'].apply(assign_value)
```

**Mr**: Es la abreviatura de "Mister" y se utiliza como un término de cortesía para dirigirse a un hombre, independientemente de su estado civil.

**Mrs:** Es la abreviatura de "Missus" y se utiliza para dirigirse a una mujer casada. También se puede utilizar para referirse a una mujer divorciada o viuda que no prefiera usar "Ms."

**Miss:** Se utiliza para dirigirse a una mujer soltera o una niña joven. Generalmente, se utiliza con el apellido de la persona.

**Master:** Se utiliza para dirigirse a un niño joven. Históricamente, se usaba para denotar a un niño que aún no había alcanzado la edad adulta.

**Major:** Es un rango militar utilizado en varios ejércitos y fuerzas armadas, que generalmente está por encima de un capitán pero por debajo de un teniente coronel.

**Rev:** Es la abreviatura de "Reverend" y se utiliza para referirse a un clérigo, como un pastor, un sacerdote o un ministro religioso.

**Dr:** Es la abreviatura de "Doctor" y puede referirse a alguien que tiene un título académico de doctorado (Ph.D.) o a un médico con formación en medicina
